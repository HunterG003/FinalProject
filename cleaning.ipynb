```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('./Data/TestData.csv')
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df.info()
```

```python
df.plot.scatter(x='city_fuel_economy', y='year')
```

```python
df['year'].value_counts().sort_values(ascending=True).plot(kind='barh', figsize=(20,20))
```

```python
df['body_type'].value_counts().plot(kind='bar', figsize=(20, 10))
```

```python
df.loc[df['horsepower'] == 'False', 'horsepower'] = np.NaN
df['horsepower'] = df['horsepower'].astype('float64')
df.loc[df['city_fuel_economy'] == 'False', 'city_fuel_economy'] = np.NaN
df.loc[df['engine_cylinders'] == 'False', 'engine_cylinders'] = np.NaN
df.loc[df['engine_displacement'] == 'False', 'engine_displacement'] = np.NaN
df.loc[df['body_type'] == 'False', 'body_type'] = np.NaN
df.loc[df['mileage'] == 0.0, 'mileage'] = np.NaN
df.loc[df['owner_count'] == 0.0, 'owner_count'] = np.NaN
```

```python
suv_df = df[df['body_type'] == 'SUV / Crossover']
```
